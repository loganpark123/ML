```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # statistical data visualization
%matplotlib inline
```

```python

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train.dropna()
test.dropna()
```

```python
train['satisfaction'] = train['satisfaction'].replace(to_replace='neutral or dissatisfied', value=0)
train['satisfaction'] = train['satisfaction'].replace(to_replace='satisfied', value=1)
train = pd.get_dummies(train)
train.iloc[:,5:].head(5)
```

```python
train.shape
train
```

```python
train_labels = np.array(train['satisfaction'])
train_features = train.drop(labels=['satisfaction'], axis=1)
train_feature_list = list(train.columns)
```

```python
train = np.array(train)
```

```python
test['satisfaction'] = test['satisfaction'].replace(to_replace='neutral or dissatisfied', value=0)
test['satisfaction'] = test['satisfaction'].replace(to_replace='satisfied', value=1)
test = pd.get_dummies(test)
test.iloc[:,5:].head(5)
test_labels = np.array(test['satisfaction'])
test_features = test.drop(labels=['satisfaction'], axis=1)
test_feature_list = list(test.columns)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
train_features = train_features.fillna(0)
test_features = test_features.fillna(0)
```

```python
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
```

```python
predictions = rf.predict(test_features)
predictions = np.round(predictions)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print("Confusion Matrix:")
print(confusion_matrix(test_labels,predictions))
print("Classification Report:")
print(classification_report(test_labels,predictions))
```
